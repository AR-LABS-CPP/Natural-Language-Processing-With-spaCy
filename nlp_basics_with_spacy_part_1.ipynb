## What is NLP?
According to Wikipedia,
"Natural Language Processing (NLP)" is an area of computer science and artificial intelligence concerned with the interactions between computers and human (natural) languages, in particular how to program computers to process and analyze large amounts of natural langauge data.

- Computers are very good at handling and performing operations on numerical information such as sales numbers, physical measurements, quantifiable categories. But what to do when we have "text data"?

- For a human identifying information from text documents is very easy but a computer needs specialized processing techniques in order to "understand" raw text data because text data is highly unstructured and can be in multiple languages e.g. English, French, German (This notebook deals with the english langauge).

- Natural Language Processing attempts to use a variety of techniques in order to create structure out of text data.

### Example Use Cases of NLP:
- Classifying Emails as Spam vs Legitimate
- Sentiment Analysis of a chunk of text
- Trend Analysis from feedback forms
- Understanding text commands, "Hey siri, play this song"

## Stemming

Stemming is one stage in text mining pipelines that converts raw text data into a structured format for machine processing. It is often used in search engines and other information retrieval systems. Stemming is also used in preparing text data for deep learning tools, such as large language models, and various NLP tasks, such as sentiment analysis and text classification.

Stemmers are the algorithms used to reduce different forms of a word to a base form. Essentially, they do this by removing specific character strings from the end of word tokens. Stemmers thus do not account for prefixes. Most stemmers contain a list of common language-specific suffixes against which the algorithm matches input word tokens. If the algorithm matches a word to one of the suffixes, and stripping the suffix does not violate pre-specified rules in the algorithm (e.g. character length restrictions), then the algorithm removes the suffix from the word.

Stemming algorithms differ in a number of ways. One primary difference is in the conditions or rules that determine whether to strip a given suffix from a token. Additionally, some stemmers have processes to correct for malformed stem words, and so they limit under-stemming and over-stemming.

Spacy does not include stemming by default due to which we will shift gears towards NTLK and look into Porter Stemmer and the Snowball Stemmer.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
words = ["run", "runner", "ran", "runs", "easily", "fairly", "fairness"]
```

```python
# stem the words
for word in words:
    print(f"Original word: {word}, Stemmed word: {porter_stemmer.stem(word)}")
```

```python
# stem the words using snowball stemmer
for word in words:
    print(f"Original word: {word}, Stemmed word: {snowball_stemmer.stem(word)}")
```

```python
words = ["generous", "generation", "generously", "generate"]
```

```python
for word in words:
    print(f"Original word: {word}, Stemmed word: {snowball_stemmer.stem(word)}")
```

## Lemmatization

According to wikipedia: Lemmatization (or less commonly lemmatisation) in linguistics is the process of grouping together the inflected forms of a word so they can be analysed as a single item, identified by the word's lemma, or dictionary form.

In computational linguistics, lemmatization is the algorithmic process of determining the `lemma` of a word based on its intended meaning. Unlike stemming, lemmatization depends on correctly identifying the intended part of speech and meaning of a word in a sentence, as well as within the larger context surrounding that sentence, such as neighbouring sentences or even an entire document.

In many languages, words appear in several inflected forms. For example, in English, the verb `to walk` may appear as `walk`, `walked`, `walks` or `walking`. The base form, 'walk', that one might look up in a dictionary, is called the `lemma` for the word. The association of the base form with a part of speech is often called a `lexeme` of the word.

## Phrase Matching and Vocabulary

Phrase matching can be thought of as a powerful version of Regular Expressions where we actually take parts of speech into account for our pattern search.

#### Rule based matching

```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(NLP.vocab)
```

```python
# pattern_one: SolarPower
# pattern_two: Solar-Power
# pattern_three: Solar Power

pattern_one = [
    { "LOWER": "solarpower" }
]

pattern_two = [
    { "LOWER": "solar" },
    { "IS_PUNCT": True },
    { "LOWER": "power" }
]

pattern_three = [
    { "LOWER": "solar" },
    { "LOWER": "power" }
]
```

Let's break this down:
* `pattern1` looks for a single token whose lowercase text reads 'solarpower'
* `pattern2` looks for two adjacent tokens that read 'solar' and 'power' in that order
* `pattern3` looks for three adjacent tokens, with a middle token that can be any punctuation.<font color=green>*</font>

```python
matcher.add(
    "SolarPower",
    [
        pattern_one,
        pattern_two,
        pattern_three
    ]
)
```

```python
doc = NLP(u"The Solar Power industry is growing rapidly as people are purchasing solar-power equipment")
```

```python
matches = matcher(doc)
```

```python
for idx, (_, start, end) in enumerate(matches):
    print(f"Match {idx}: {doc[start:end]}")
```

#### Setting pattern options and quantifiers

You can make token rules optional by passing an `'OP':'*'` argument. This lets us streamline our patterns list:

```python
# pattern_one: SolarPower
# pattern_two: Solar-Power
# pattern_three: Solar Power

pattern_one = [
    { "LOWER": "solarpower" }
]

pattern_two = [
    { "LOWER": "solar" },
    { "IS_PUNCT": True, "OP": "*" },
    { "LOWER": "power" }
]
```

```python
matcher.remove("SolarPower")
```

```python
matcher.add(
    "SolarPower",
    [
        pattern_one,
        pattern_two,
    ]
)
```

```python
doc = NLP(u"The Solar Power industry is growing rapidly as people are purchasing solar---power equipment")
```

```python
matches = matcher(doc)
```

```python
for idx, (_, start, end) in enumerate(matches):
    print(f"Match {idx}: {doc[start:end]}")
```

This found both two-word patterns, with and without the hyphen!

The following quantifiers can be passed to the `'OP'` key:

<tr ><td><span >\+</span></td><td>Require the pattern to match 1 or more times</td></tr>

#### Phrase Matching

```python
from spacy.matcher import PhraseMatcher
```

```python
matcher = PhraseMatcher(NLP.vocab)
```

```python
with open("./match_example_A.txt") as file:
    new_document = NLP(file.read())
```

```python
phrases = ["boy looked", "deer stopped", "went home", "frog home"]
```

```python
phrase_patterns = [NLP(phrase) for phrase in phrases]
```

```python
matcher.add("PhraseMatcher", phrase_patterns)
```

```python
matches = matcher(new_document)
```

```python
for idx, (_, start, end) in enumerate(matches):
    print(f"Match {idx}: {new_document[start:end]}")
```
